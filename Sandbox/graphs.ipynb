```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
```

```python
G.add_edges_from([(1,2),(1,3)])
G.add_node(1)
G.add_edge(1,2)
G.add_node("spam")       # adds node "spam"
G.add_nodes_from("spam") # adds 4 nodes: 's', 'p', 'a', 'm'
```

```python
nx.draw(G)
```

```python
nx.draw_networkx(G)
```

```python
N = nx.watts_strogatz_graph(100, 6, 0)
nx.draw_kamada_kawai(N)
```

```python
N = nx.random_regular_graph(6, 100)
nx.draw_kamada_kawai(N)
```

```python
N = nx.relaxed_caveman_graph(50, 4, 0.5)
nx.draw_kamada_kawai(N)
```

```python
N = nx.grid_2d_graph(10, 10)
N.add_edge((0,0), (1,2))
N.add_edge((4,3), (1,2))
N.add_edge((7,7), (8,8))
N.add_edge((6,6), (4,7))
nx.draw_kamada_kawai(N)
```
